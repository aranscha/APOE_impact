```python
df_rawcounts = pd.read_csv(rawcounts_path, sep='\t')
df_rawcounts.index = df_rawcounts.gene
df_rawcounts = df_rawcounts.drop(columns=['gene'])
df_rawcounts = df_rawcounts.T
```

```python
import PyWGCNA # do not forget to restart runtime after pip install, otherwise the import could fail!

pywgcna = PyWGCNA.WGCNA(geneExp=df_rawcounts)
```
